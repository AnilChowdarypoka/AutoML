```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import warnings
warnings.filterwarnings("ignore")
```

```python
Numeric_columns = []
Dropped_columns = []
columns = []
```

```python

def encoding(data) :
       # print("Dropped columns are : ")
        for i in data.columns:
            if(data[i].dtype == np.int32 or data[i].dtype == np.int64 or data[i].dtype== np.float32 or data[i].dtype== np.float64 ):
                pass
            else:
                data[i] = data[i].astype("category").cat.codes

def cleaning1(data):
        
        for i in data.columns:
            if((((data[i].isnull().sum())/len(data[i]))*100) > 50):
                data.drop(i,axis = 1,inplace = True)
                Dropped_columns.append(i)
                #print(i)
        
            else:
                if data[i].dtype == np.int32 or data[i].dtype == np.int64 or data[i].dtype == np.float32 or data[i].dtype == np.float64:
                    Numeric_columns.append(i)
                    data[i] = data[i].fillna(data[i].mean())
                else:
                    mode = data[i].mode()
                    data[i] = data[i].fillna(mode.loc[0])
def dealing_with_inconsistentdata(data):
    for i in Numeric_columns:
            Q1 =data[i].quantile(0.25)
            Q3 =data[i].quantile(0.75)
            IQR = Q3 - Q1
            lower = Q1 - (IQR*1.5)
            Upper = Q3 + (IQR*1.5)
            data[i] = data[i].apply(lambda x : x if(x > lower and x < Upper) else None)
    cleaning1(data)
    #print("Total Number of columns Dropped are : ", len(Dropped_columns))
        
```

```python
class train_test_split1:
    
    
    def __init__ (self,data,y):
        self.data = data
        self.y = y
        #self.train_size = 0.8
      
    def train_test_split(self):
        
            print("Our Target Column is" , y)
            print()
            train_size = float(input("Enter train size value for spliting the dataset with in the range of (0-1) : "))
            if(train_size > 0 and train_size < 1):
                pass
            else:
                print()
                print("Size Must Lies between 0 - 1 ")
               
            
            data_len = len(self.data[y])
            Training= random.sample(range(0,data_len),int(((train_size)*data_len)))
            Testing = []
            for i in range(0,data_len):
                if i in Training:
                    pass
                else:
                    Testing.append(i)

    #print(len(Testing))
    #print(len(Training))
            Training_data = data.loc[Training,:]

            Testing_data = data.loc[Testing,:]
    #print(Training_data.shape)
    #print(Testing_data.shape)
            X_train = Training_data.drop(y,axis = 1)
            X_test = Testing_data.drop(y, axis = 1)
            y_train = Training_data[y]
            y_test = Testing_data[y]
            return X_train,X_test, y_train,y_test
        
```

```python
class Regression:
    
    def __init__ (self,X_train, X_test, y_train,y_test):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test
    
    def Linear_Regression(self):
        from sklearn.linear_model import LinearRegression
        model = LinearRegression()
        model.fit(X_train,y_train)
        predicted = model.predict(X_test)
        score = model.score(X_test,y_test)
        return score 

    def SVR(self):
    
        from sklearn.svm import SVR
        c = int(input("Enter value of C for SVR : "))
        ker = input("Choose Linear, Polynomial, RBF and Sigmoid for SVR : ")
        model = SVR(gamma = "auto",C=c,kernel= ker)
        model.fit(X_train,y_train)
        predicted = model.predict(X_test)
        score = model.score(X_test,y_test)
        return score 
    def Random_forestr(self):
   
        from sklearn.ensemble import RandomForestRegressor
        no_of_sample_trees = int(input("Enter no of sample Trees for Random Forest "))
        model = RandomForestRegressor(n_estimators = no_of_sample_trees)
        model.fit(X_train,y_train)
        predicted = model.predict(X_test)
        score2 = model.score(X_test,y_test)
        return score2
    def polynomial(self):
        from sklearn.linear_model import LinearRegression
        from sklearn.preprocessing import PolynomialFeatures
        model = LinearRegression()
        degree1 = int(input("Enter the degree for the polynomial Regression : "))
        poly = PolynomialFeatures(degree = degree1)
        X_train_pr = poly.fit_transform(X_train)
        X_test_pr = poly.fit_transform(X_test)
        model.fit(X_train_pr,y_train)
        predicted = model.predict(X_test_pr)
        score = model.score(X_test_pr,y_test)
        return score 
```

```python
class classification:
    
    def __init__ (self,X_train, X_test, y_train,y_test):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test
        
    def SVC(self):
        from sklearn.svm import SVC
        c = int(input("Enter value of c for SVC : "))
        ker = input("Choose Linear, Polynomial, RBF and Sigmoid for SVC : ")
        model = SVC(gamma = "auto",C=c,kernel= ker)
        model.fit(self.X_train,self.y_train)
        predicted = model.predict(self.X_test)
        score1 = model.score(self.X_test,self.y_test)
        return score1

    def Logistic_Regression(self):
        from sklearn.linear_model import LogisticRegression
        model = LogisticRegression()
        model.fit(self.X_train,self.y_train)
        predicted = model.predict(self.X_test)
        score2 = model.score(self.X_test,self.y_test)
        return score2

    def Decision_tree(self):
        from sklearn import tree
        model = tree.DecisionTreeClassifier()
        model.fit(self.X_train,self.y_train)
        predicted = model.predict(self.X_test)
        score2 = model.score(self.X_test,self.y_test)
        return score2
    def Random_forestc(self):
        from sklearn.ensemble import RandomForestClassifier
        no_of_sample_trees = int(input("Enter no of sample Trees for Random Forest "))
        model = RandomForestClassifier(n_estimators = no_of_sample_trees)
        model.fit(self.X_train,self.y_train)
        predicted = model.predict(self.X_test)
        score2 = model.score(self.X_test,self.y_test)
        return score2
```

```python
class Model_selection:
    
    def regression(self):
        reg = Regression(X_train, X_test, y_train,y_test)
        score1  = reg.Linear_Regression()
        score2  = reg.SVR()
        score3 = reg.Random_forestr()
        score4 = reg.polynomial()
        Top = max(score1,score2,score3,score4)
        if(Top == score1):
            return (" Linear Regression and accuracy is " , Top)
        elif(Top == score2):
            return (" SVM and accuracy is ",Top)
        elif(Top == score4):
            return (" Polynomial Regression and accuracy is ",Top)
            
        else:
            return ("Random Forest and accuracy is ", Top)
    def classificaion(self):
        Clsf = classification(X_train, X_test, y_train,y_test)
        score1  = Clsf.Logistic_Regression()
        score2  = Clsf.SVC()
        score3 = Clsf.Decision_tree()
        score4 = Clsf.Random_forestc()
        Top = max(score1,score2,score3,score4)
        if(Top == score1):
            return (" Logistic Regression and accuracy is " , Top)
        elif(Top == score2):
            return (" SVC and accuracy is " , Top)
        elif(Top == score3):
            return (" Decision Tree and accuracy is ",Top)
        else:
            return ("Random Forest and accuracy is ", Top)
```
